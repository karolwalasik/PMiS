```python
import math
import numpy as np
from numpy import sin, cos
from scipy.integrate import odeint
from matplotlib import pyplot as plt
```

### Funkcje

---



```python
def equations(y0,t):
  fi, x = y0
  f = [x,-(g/l)*sin(fi)]
  return f

def plot_results(time,fi1,fi2,fi0):
  plt.plot(time,fi1[:,0])
  plt.plot(time,fi2)

  s= 'Kat poczatkowy = ' + str(fi0) + ' stopni'
  plt.title('Wahadło matematyczne: ' + s)
  plt.xlabel('time(s)')
  plt.ylabel('angle(rad)')
  plt.grid(True)
  plt.legend(['nieliniowe','liniowe'],loc='lower right')
  plt.show()
```

###Parametry

```python
g=9.81
l=0.716
# kat 0.072

time = np.arange(0,10.0,0.025)
```

###Warunki poczatkowe

```python
initial_angle = 0.072
fi0 = np.radians(initial_angle)
# x0 - predkosc
x0 = np.radians(0.0)
```

###Znalezienie rozwiązania równanie nieliniowego

```python
fi1 = odeint(equations,[fi0,x0],time)
print(fi1)
```

###Znalezienie rozwiązania równania liniowego

```python
w = np.sqrt(g/l)
fi2 = [fi0 * cos(w*t) for t in time]
# czestotliwosc
f = (1/(2*math.pi)) * w
print(f)
#okres
T =1/f
print(T)
print(len(fi2))
timeT = np.arange(0,T,T/399.5)
fi2 = [fi0 * cos(w*t) for t in timeT]
plot_results(timeT,fi1,fi2,0.07)
```

###Wykres

```python
plot_results(time,fi1,fi2,initial_angle)
```
